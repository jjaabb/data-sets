```python
categorys = []
for index, row in df18.iterrows():
    #print(row['category'])
    if row['category'] not in categorys:
        categorys.append(row['category'])
    else:
        pass 
    
print(categorys)
```

```python
cat = {}
for index, row in df18.iterrows():
    #print(row['category'])
    if row['category'] not in cat.keys():
        cat[row['category']] = 1
        
    else:
        cat[row['category']] += 1        
```

```python
from collections import Counter
```

```python
category_counter = Counter(cat)
category_counter['Poetry']
```

```python
most_comon_categorys = category_counter.most_common()[:20]
most_comon_categorys
```

```python
# Calculate % of top 20 categorys of 159 all existing categorys
top_20 = 0  
for item in most_comon_categorys:
    for i in item:
        if type(i) == int:
            top_20 += int(i)
        else:
            pass
print(round(top_20 / 378661 * 100, ndigits=2))
```
